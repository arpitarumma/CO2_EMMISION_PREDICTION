```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
df=pd.read_csv('/content/drive/MyDrive/DATASETS/owid-co2-data.csv')
```

```python
df.columns
```

```python
df.shape
```

```python
df.duplicated().sum()
```

```python
null=df.isnull().sum()
```

```python
drop=[]
n = 0
for i in null:
  if i > 15000:
    col = null.index[n]
    drop.append(col)
  n+=1
drop
```

```python
df.drop(drop,axis=1,inplace=True)
```

```python
df.columns
```

```python
co2_con=df.dropna()
fig=px.histogram(co2_con,x='country',y='co2_per_capita',histfunc='avg',color='country')
fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()
```

```python
co2_ind = df.dropna()
co2_ind = co2_ind[co2_ind['country'] == 'India']
fig = px.scatter(co2_ind, x='year', y='co2_per_capita', color='co2_per_capita', title="INDIA CO2 PER CAPITA") # Changed 'tile' to 'title' to set the plot title.

fig.show()
```

```python
co2_ind = df.dropna()
co2_ind = co2_ind[co2_ind['country'] == 'Afghanistan']
fig = px.scatter(co2_ind, x='year', y='co2', color='co2')
fig.show()
```
